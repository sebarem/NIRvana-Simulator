```python
img = zeros((512,512),dtype=int32)
```

```python
ReadNoise = 120 #e
Signal = 1.0 #signal over read noise
ExpTime = 3
DarkCurrentNIRvana = 300 #e/p/s
DarkCurrentNIRvanaST = 1500 #e/p/s

ShotNoise = sqrt(Signal*ReadNoise)
DarkNoiseNIRvana = sqrt(DarkCurrentNIRvana*ExpTime)
DarkNoiseNIRvanaST = sqrt(DarkCurrentNIRvanaST*ExpTime)
print "DC + Signal NIRvana = ", DarkCurrentNIRvana*ExpTime+Signal*ReadNoise
print "DC + Signal NIRvanaST = ", DarkCurrentNIRvanaST*ExpTime+Signal*ReadNoise
```

```python
img[104:115,205:306] = Signal*ReadNoise
img[124:135,205:306] = Signal*ReadNoise
img[144:155,205:306] = Signal*ReadNoise

img[360:371,205:306] = Signal*ReadNoise
img[380:391,205:306] = Signal*ReadNoise
img[400:411,205:306] = Signal*ReadNoise
```

```python
NoiseNIRvana = sqrt(ReadNoise**2+ShotNoise**2+DarkNoiseNIRvana**2)
NoiseNIRvanaST = sqrt(ReadNoise**2+ShotNoise**2+DarkNoiseNIRvanaST**2)
img[0:256,:] = img[0:256,:] + NoiseNIRvana*randn(256,512)
img[256:512,:] = img[256:512,:] + NoiseNIRvanaST*randn(256,512)
print img.max()
print "Total Signal = ", Signal*ReadNoise
print "Noise of NIRvana = ", NoiseNIRvana.mean(), "S/N = ", Signal*ReadNoise/NoiseNIRvana
print "Noise of NIRvanaST = ",NoiseNIRvanaST.mean(), "S/N = ", Signal*ReadNoise/NoiseNIRvanaST
```
